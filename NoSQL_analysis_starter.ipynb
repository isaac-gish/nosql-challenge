### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = [{'$match': {'scores.Hygiene': 20}}]

results = list(establishments.aggregate(query))
# Use count_documents to display the number of documents in the result
count = len(results)
print(f'The number of documents: {count}')
# Display the first document in the results using pprint
pprint(results[0])
```

```python
import pandas as pd
# Convert the result to a Pandas DataFrame
hygiene_df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
rows = len(hygiene_df)
print(f'Nubmer of rows: {rows}')
# Display the first 10 rows of the DataFrame
hygiene_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# # Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = [{'$match' : {'LocalAuthorityName': {'$regex' :'London'}, 'RatingValue': {'$gte': 4}}}]

london_results = list(establishments.aggregate(query))
# Use count_documents to display the number of documents in the result
count = len(london_results)
print(f'The number of documents: {count}')
# Display the first document in the results using pprint
pprint(london_results)

```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score



degree_search = 0.01

#Calculate min and max latitudes
latitude = establishments.find({'BusinessName':"Penang Flavours"},{'geocode.latitude':1}).limit(1)
latitude = float(latitude[0].get('geocode').get('latitude').to_decimal())
min_lat = latitude - degree_search
max_lat = latitude + degree_search


#Calculate min and max longitude
longitude = establishments.find({'BusinessName':"Penang Flavours"},{'geocode.longitude':1}).limit(1)
longitude = float(longitude[0].get('geocode').get('longitude').to_decimal())
min_long = longitude - degree_search
max_long = longitude + degree_search

#Create query
query = {'$and':[ {'geocode.latitude': {'$gte': min_lat }}, {'geocode.latitude': {'$lte': max_lat }},
                 {'geocode.longitude': {'$gte': min_long }},{'geocode.longitude': {'$lte': max_long }},{'RatingValue':{'$eq': 5}}
                 ]}

#Sort by Highest to Lowest Rating Value
sort =  [('RatingValue', 1)]

#Limit to the top 5
limit = 5

# Print the results
results = list(establishments.find(query).sort(sort).limit(limit))
pprint(results)
```

```python
# Convert result to Pandas DataFrame
close_df = pd.DataFrame(results)
close_rows = len(close_df)
print(f'Nubmer of rows: {close_rows}')
# Display the first 10 rows of the DataFrame
hygiene_df.head(10)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

#Find only hygiene scores equal to 0
match = {'$match': {'scores.Hygiene': 0}}
#Group by local authority
group = {'$group' : {'_id' : '$LocalAuthorityName', 'count': {'$sum' : 1}}}
#Sort by descending order (-1)
sort = {'$sort': { 'count': -1}}

pipeline = [match, group, sort]
results = list(establishments.aggregate(pipeline))
# Print the number of documents in the result
print('Number of documents:', len(results))
# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
gross_df = pd.DataFrame(results)

rows = len(gross_df)
# Display the number of rows in the DataFrame
print(f"Number of rows: {rows}")
# Display the first 10 rows of the DataFrame
gross_df.head(10)
```
